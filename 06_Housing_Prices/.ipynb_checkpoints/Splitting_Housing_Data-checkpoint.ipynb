# Loading Data

```scala
val housing = spark.read.format("csv").option("header","true").option("inferSchema","true").load("housing.csv")
housing.show(10)
```

```scala
housing.printSchema
```

```scala
housing.summary().show()
```

```scala
housing.createOrReplaceTempView("housing")
```

```scala
%%python

import databricks.koalas as ks

housing = spark.sql("select * from housing").to_koalas()
print(housing.head())
```

```scala
%%python
import tempfile
from matplotlib import pyplot as plt
from IPython.display import Image
```

```scala
%%python
housing[['longitude']].plot.hist(bins=40, alpha=0.5)
with tempfile.NamedTemporaryFile(suffix=".png") as fo:
    plt.savefig(fo.name)
    retval = Image(filename=fo.name)
```
