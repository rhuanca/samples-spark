```scala
val housing = spark.read.format("csv").option("header","true").option("inferSchema","true").load("housing.csv")
housing.createOrReplaceTempView("housing")
housing.show(10)
```

```scala
%%python
import databricks.koalas as ks
import tempfile
from matplotlib import pyplot as plt
from IPython.display import Image
ks.set_option("plotting.max_rows", 1000000)
```

# Stratified Sampling

```scala
import org.apache.spark.ml.feature.Bucketizer

val bucketizer = new Bucketizer()
  .setInputCol("median_income")
  .setOutputCol("income_cat")
  .setSplits(Array(0,1.5,3.0,4.5,6,Double.PositiveInfinity))

val housing_temp = bucketizer.transform(housing)
housing_temp.show(10)
```

```scala
housing_temp.groupBy("income_cat").count.show
```

```scala
val test = housing_temp.stat.sampleBy("income_cat", Map(0 -> 0.2, 1->0.2, 2->0.2, 3->0.2, 4->0.2, 5->0.2), 47)
val train = housing_temp.except(test)

train.createOrReplaceTempView("train")
```

```scala
%%python
import numpy as np
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colors import to_rgba

train = spark.table("train").to_koalas()
train.plot.scatter(x="longitude", y="latitude",c=to_rgba('b', 0.1))

with tempfile.NamedTemporaryFile(suffix=".png") as fo:
    plt.savefig(fo.name)
    retval = Image(filename=fo.name)
```

# Correlation

```scala
%%python

train = spark.table("train").toPandas()
corr_matrix = train.corr()
print(corr_matrix)
```

```scala
%%python
print(corr_matrix["median_house_value"].sort_values(ascending=False))
```

```scala

import org.apache.spark.ml.feature.VectorAssembler
import org.apache.spark.ml.linalg.Vectors

val assembler = new VectorAssembler()
  .setInputCols(Array("longitude", "latitude", "housing_median_age", 
                      "total_rooms", "total_bedrooms", "population", "households", 
                      "median_income"))
  .setOutputCol("features")

val train_tmp = assembler.transform(train.where($"total_bedrooms".isNotNull))
    .drop("ocean_proximity")
    .drop("income_cat")

train_tmp.show(10)
train_tmp.createOrReplaceTempView("train_tmp")
```

```scala
import org.apache.spark.ml.linalg.{Matrix, Vectors}
import org.apache.spark.ml.stat.Correlation
import org.apache.spark.sql.Row
import org.apache.spark.sql.Dataset

val Row(coeff1: Matrix)  = Correlation.corr(train_tmp, "features").head
println(s"Pearson correlation matrix:\n $coeff1")

val Row(coeff2: Matrix) = Correlation.corr(train_tmp, "features", "spearman").head
println(s"Spearman correlation matrix:\n $coeff2")
```

```scala
%%python
train = spark.sql("""
select * from train
where total_bedrooms is not null
""").to_koalas()

print("==> pearson correlation")
print(train.corr()["median_house_value"].sort_values(ascending=False))

print("==> spearman correlation")
print(train.corr("spearman")["median_house_value"].sort_values(ascending=False))
```

```scala
%%python
from pandas.plotting import scatter_matrix
attributes = ["median_house_value","median_income","total_rooms","housing_median_age"]

train = spark.table("train").toPandas()

scatter_matrix(train[attributes], figsize=(16,14))

with tempfile.NamedTemporaryFile(suffix=".png") as fo:
    plt.savefig(fo.name)
    retval = Image(filename=fo.name)
```

```scala
%%python
import seaborn as sns

train = spark.table("train").to_koalas()
attributes = ["median_house_value","median_income","total_rooms","housing_median_age"]

corr_values = train[attributes].corr().to_numpy()

sns.heatmap(corr_values, cmap='Blues', annot = True, square = False,
            xticklabels=attributes, yticklabels=attributes)

with tempfile.NamedTemporaryFile(suffix=".png") as fo:
    plt.savefig(fo.name)
    retval = Image(filename=fo.name)
```
